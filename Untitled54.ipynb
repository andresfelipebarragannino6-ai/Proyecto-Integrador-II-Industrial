```python
import numpy as np
from scipy.integrate import odeint
from scipy.optimize import minimize

# --- 1. EL MOTOR DE LA FÍSICA (MODELO DINÁMICO) ---
def maillard_physics(y, t, k1, k2):
    P, I, M = y
    return [-k1*y[0], k1*y[0]-k2*y[1], k2*y[1]]

# --- 2. FUNCIÓN DE PÉRDIDA DE LA PINN (PHYSICS-INFORMED LOSS) ---
def pinn_loss(params, t_obs, all_data):
    """
    params: [lnA1, Ea1, lnA2, Ea2] -> Parámetros termodinámicos a optimizar
    all_data: Diccionario con datos experimentales a distintas Temperaturas
    """
    lnA1, Ea1, lnA2, Ea2 = params
    R = 8.314 / 1000 # Constante en kJ/(mol·K)
    total_loss = 0

    for T_c in [200, 240]: # Entrenamiento multitérmico
        T_k = T_c + 273.15

        # RESTRICCIÓN FÍSICA: Cálculo de k mediante Arrhenius
        k1 = np.exp(lnA1) * np.exp(-Ea1 / (R * T_k))
        k2 = np.exp(lnA2) * np.exp(-Ea2 / (R * T_k))

        # INTEGRACIÓN DE LA EDO (Física del proceso)
        y0 = all_data[T_c]['obs'][0]
        sol = odeint(maillard_physics, y0, t_obs)

        # CÁLCULO DEL ERROR PONDERADO (Data Loss + Physics Constraint)
        diff = (sol - all_data[T_c]['obs'])**2
        weighted_diff = diff / (all_data[T_c]['err']**2) # Informed Weighting
        total_loss += np.sum(weighted_diff)

    return total_loss

# --- 3. ENTRENAMIENTO (OPTIMIZACIÓN DE MEMORIA L-BFGS-B) ---
initial_guess = [10, 50, 5, 30] # Semilla inicial para lnA y Ea
res_pinn = minimize(pinn_loss, initial_guess, args=(t_obs, all_data),
                    method='L-BFGS-B')

# Extracción de parámetros finales optimizados por la red
lnA1_f, Ea1_f, lnA2_f, Ea2_f = res_pinn.x
```
